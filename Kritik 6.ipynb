```python
import matplotlib.pyplot as plt
import numpy as np

def normal_density(mean, variance, x):
    return (1 / np.sqrt(2 * np.pi * variance)) * np.exp(-((x - mean)**2) / (2 * variance))

x_values = np.linspace(-10, 10, 1000)

# Plot for different values of mu and sigma^2
plt.plot(x_values, normal_density(0, 1, x_values), label="μ=0, σ²=1")
plt.plot(x_values, normal_density(0, 2, x_values), label="μ=0, σ²=2")
plt.plot(x_values, normal_density(2, 1, x_values), label="μ=2, σ²=1")
plt.xlabel("x")
plt.ylabel("f_X(x)")
plt.title("Normal Distribution PDF")
plt.legend()
plt.show()
```

```python
def integrate_normal(mean, variance, a, b, n=1000):
    x_values = np.linspace(a, b, n)
    y_values = normal_density(mean, variance, x_values)
    dx = (b - a) / n
    integral = np.sum(y_values) * dx
    return integral
```

```python
def expected_uniform(a, b):
    return (a + b) / 2

a, b = 2, 10
expected_value = expected_uniform(a, b)
print(f"Expected value of X ~ U({a}, {b}): {expected_value}")
```
