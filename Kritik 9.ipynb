```python
# case 1
import numpy as np

def roots(f, a, b):
    if f(a) * f(b) >= 0:
        print("No root in interval")
        return None
        
    while (b - a) / 2 > 10**-10:
        mid = (a + b) / 2
        if f(mid) == 0:
            return mid
        elif f(a) * f(mid) < 0:
            b = mid
        else:
            a = mid

    return (a + b) / 2

# case 1
def f1(x):
    return np.exp(x)+np.log(x)

root = roots(f1, 0.001, 1)
print (root)

# case 2 
def f2(x):
    return np.arctan(x) - x**2

root = roots(f2, 0.001, 2)
print(root)

# case 3
def f3(x):
    return np.sin(x)/np.log(x)

root = roots(f3, 3, 4)
print(root)

# case 4
def f4(x):
    return np.log(np.cos(x))

root = roots(f4, 5, 7)
print(root)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a range of x values
x_vals = np.linspace(-2, 1.01, 400)

# Compute y values for the first equation y = x^5 - 4x^3
y_vals_1 = x_vals**5 - 4*x_vals**3

# Create a range of y values for the second equation x = y^5 - 4y^2
y_vals_2 = np.linspace(-2, 1.01, 400)
x_vals_2 = y_vals_2**5 - 4*y_vals_2**2

# Create the plot
plt.figure(figsize=(8, 8))

# Plot the first curve y = x^5 - 4x^3
plt.plot(x_vals, y_vals_1, label=r'$y = x^5 - 4x^3$', color='b')

# Plot the second curve x = y^5 - 4y^2
plt.plot(x_vals_2, y_vals_2, label=r'$x = y^5 - 4y^2$', color='r')

# Add labels and title
plt.title('Graphs of $y = x^5 - 4x^3$ and $x = y^5 - 4y^2$')
plt.xlabel('x')
plt.ylabel('y')

# Set aspect ratio to be equal to ensure proper visualization
plt.gca().set_aspect('equal', adjustable='box')

# Show the grid and legend
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def gradient_descent(f, learning_rate, initial_point, epsilon=1e-5, max_iterations=10000):
    # Define the derivative function using symmetric difference approximation
    def deriv(f, base_point, h=1e-10):
        return (f(base_point + h) - f(base_point - h)) / (2 * h)

    # Lists to store the sequence of x and f(x)
    x_coords = [initial_point]
    y_coords = [f(initial_point)]

    # Start gradient descent loop
    current_x = initial_point
    for _ in range(max_iterations):
        grad = deriv(f, current_x)  # Calculate the derivative at current_x
        new_x = current_x - learning_rate * grad  # Gradient descent update rule

        # Append the new point and the function value at that point
        x_coords.append(new_x)
        y_coords.append(f(new_x))

        # Check the stopping condition
        if abs(grad) < epsilon:
            break

        # Update current_x for the next iteration
        current_x = new_x

    # Plotting the function and the sequence of points
    plot_range = np.linspace(min(x_coords)-0.5, max(x_coords)+0.5, 1000)
    function_range = [f(x) for x in plot_range]

    # Plot the function
    plt.plot(plot_range, function_range, label='f(x)', color='blue')
    
    # Plot the gradient descent steps (sequence of points)
    plt.plot(x_coords, y_coords, 'ro-', label='Gradient Descent Steps')

    # Labeling the plot
    plt.title('Gradient Descent for Function Minimization')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Return the final x and f(x)
    return round(x_coords[-1], 3), round(y_coords[-1], 3)

# Example usage: Minimize the function f(x) = x^2 with learning rate = 0.1, starting at x0 = 5
def f(x):
    return x ** 2

# Running the gradient descent function
final_x, final_f_x = gradient_descent(f, learning_rate=0.1, initial_point=5)

print(f"Final x: {final_x}, f(x): {final_f_x}")
```
